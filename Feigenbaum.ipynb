```python
#! /usr/bin/python3
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.pyplot import figure 

import random
from pylab import plot,show
import warnings
warnings.filterwarnings("ignore")


####################################
#### MAPA LOGISTICO DE FEIGENBAUM 

#declaracion de variables iniciales
r=1       #cte
k=0       #inicializacion iteraciones
param_x=[0.45, 0.55, 0.05]

#inicializando listas de almacenamiento
X=[]
Y=[]

#lista de formas de puntos (para poder diferencias los puntos dibujados más adelante)
forma_pto=['g+','y*','k.']

#iteración con la función feigenbaum
for ix in param_x :
    while k<1000 :           
        x_k = r*(1-ix)*ix     #eq recurrencia 
        Y.append(x_k)         #almacenando
        X.append(k+1)         #comenzando en k=1
        k += 10               #k+10 para poder apreciar el decrecimiento de los puntos
        ix=x_k                #reemplazando valor obtenido para generar la recurrencia
        f_pto = random.choice(forma_pto)              #elige de manera aleatoria, de la lista "forma_pto", la forma del punto
        plot(X,Y,f_pto, linewidth=0.5)                #grafica los puntos
    
#print("##")           #flag    

plt.figure(figsize=(20,10))#usado para solo aumentar la resolución del gráfico
plt.ylabel("x= evolución")
plt.xlabel("k= # iteración")
plt.plot(X,Y,linewidth=2)                  #curva suave de fondo, contiene todos los datos
plt.show()
    
```
